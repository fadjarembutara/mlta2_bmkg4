## Visualise

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
viz1 = dataset2[['temp','pressure','humidity','wind_speed','wind_deg','clouds_all','weather_main']]
viz1
```

```python
plt.rcParams['figure.figsize'] = [14, 8]
viz1.temp()
plt.title('Distribution of temp')
plt.xlabel('temp')
plt.show()
```

```python
sns.pairplot(data=viz1, hue='weather_main')
```
